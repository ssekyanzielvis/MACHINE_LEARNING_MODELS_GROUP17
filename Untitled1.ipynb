```python
# Define bandpass filter parameters
lowcut = 0.5  # Low cutoff frequency (Hz)
highcut = 40  # High cutoff frequency (Hz)
order = 4    # Filter order

# Design the bandpass filter
nyquist = 0.5 * fs
low = lowcut / nyquist
high = highcut / nyquist
b, a = scipy.signal.butter(order, [low, high], btype='band')

# Plot the frequency response of the filter
w, h = scipy.signal.freqz(b, a, worN=8000)
frequencies = (fs * 0.5 / np.pi) * w
plt.figure(figsize=(6, 2))
plt.plot(frequencies, abs(h), 'b')
plt.title('Butterworth Filter Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid(True)
plt.show()
```

```python
# --- Signal Preprocessing: Filtering & Normalization ---
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def bandpass_filter(data, lowcut=0.5, highcut=45, fs=360, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    return filtfilt(b, a, data)
```

```python
# Extract ECG signal & Apply Filtering
ecg_signal = df['MLII'].values  # Adjust if needed
filtered_ecg = bandpass_filter(ecg_signal)

# Normalize ECG signal
normalized_ecg = (filtered_ecg - np.min(filtered_ecg)) / (np.max(filtered_ecg) - np.min(filtered_ecg))

# Plot Raw vs. Filtered ECG Signal
plt.figure(figsize=(12, 6))
plt.plot(ecg_signal[:1000], label="Raw ECG", alpha=0.5)
plt.plot(filtered_ecg[:1000], label="Filtered ECG", linewidth=1.5)
plt.title("ECG Signal - Raw vs. Filtered") 
plt.legend()
plt.show()
```

```python
# --- Detect R-peaks for Heartbeat Detection ---
peaks, _ = find_peaks(filtered_ecg, height=0.5, distance=150)

# Compute RR Intervals (Time between heartbeats)
rr_intervals = np.diff(peaks) / 360  # Convert to seconds (Sampling Rate = 360 Hz)
heart_rate = 60 / rr_intervals  # Convert RR intervals to BPM

# Plot RR Intervals
plt.figure(figsize=(10, 5))
plt.plot(rr_intervals, marker='o', linestyle='dashed')
plt.title("RR Interval Variation")
plt.xlabel("Beat Number")
plt.ylabel("RR Interval (s)")
plt.show()

# --- Heart Rate Variability (HRV) Analysis ---
print("\nðŸ”¹ HRV Analysis:")
print(f"Mean RR Interval: {np.mean(rr_intervals):.3f} sec")
print(f"STD RR Interval: {np.std(rr_intervals):.3f} sec")
print(f"Mean Heart Rate: {np.mean(heart_rate):.1f} BPM")
print(f"Heart Rate Variability: {np.std(heart_rate):.2f} BPM")
```

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt

def wavelet_transform_analysis(ecg_signal, sampling_rate):
    """
    Performs Continuous Wavelet Transform (CWT) analysis on an ECG signal and displays the results.

    Args:
        ecg_signal (numpy.ndarray): The ECG signal as a 1D numpy array.
        sampling_rate (int): The sampling rate of the ECG signal in Hz.
    """

    scales = np.arange(1, 128)  # Define the scales for the CWT

    coeffs, freqs = pywt.cwt(ecg_signal, scales, 'morl')

    plt.figure(figsize=(12, 6))

    # Calculate the time axis in milliseconds
    time_ms = np.linspace(0, len(ecg_signal) / sampling_rate * 1000, len(ecg_signal))

    plt.imshow(np.abs(coeffs), extent=[time_ms[0], time_ms[-1], scales[-1], scales[0]], cmap='jet', aspect='auto', origin='upper') # Added np.abs() for magnitude, origin, and extent
    plt.title("Continuous Wavelet Transform (CWT) of ECG")
    plt.ylabel("Frequency Scale")
    plt.xlabel("Time (ms)")  # Time in milliseconds
    plt.colorbar(label="Magnitude")

    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()



# Example usage (replace with your actual ECG signal and sampling rate)
ecg_signal = np.random.randn(1000) # Example signal - replace with your actual data
sampling_rate = 360  # Example sampling rate - replace with the correct value

wavelet_transform_analysis(ecg_signal, sampling_rate)
```

```python
def dfa(x):
    N = len(x)
    y = np.cumsum(x - np.mean(x))
    scales = np.floor(np.logspace(np.log10(10), np.log10(N/10), num=20)).astype(int)
    fluctuations = []
    
    for scale in scales:
        n_segments = int(np.floor(N / scale))
        segments = np.array_split(y[:n_segments*scale], n_segments)
        rms = []
        
        for segment in segments:
            p = np.polyfit(range(scale), segment, 1)
            fit_values = np.polyval(p, range(scale))
            rms.append(np.sqrt(np.mean((segment - fit_values)**2)))
        
        fluctuations.append(np.mean(rms))
    
    return scales, fluctuations

# Perform DFA
scales, fluctuations = dfa(rr_intervals)

# Plot DFA
plt.figure(figsize=(12, 6))
plt.loglog(scales, fluctuations, label='Detrended Fluctuation Analysis')
plt.title('Detrended Fluctuation Analysis of RR Intervals (Record 100)')
plt.xlabel('Scale')
plt.ylabel('Fluctuation')
plt.legend()
plt.show()# I'm checking for missing values in the datasets.
for file, df in dataframes.items():
    missing_values = df.isnull().sum()
    print(f"Missing values for Patient {file.split('.')[0]}:")
    print(missing_values)
    print("-"*50)

# I'm using forward fill to handle any missing values.
for file, df in dataframes.items():
    df.fillna(method='ffill', inplace=True)
```
